#### import the neccessary modules

```python
# import libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import csv
```

#### Load Airplane Data

```python
with open('data/Airplane_data.csv', mode='r', encoding='utf-8') as file:
    data = pd.read_csv(file)

data.head()
```

### Calculate the total number of accidents for each Airplane model to help calculate the accident rate

```python
# Calculate the total number of accidents for each aircraft model
accidents_per_model = data.groupby('Model')['Event.Id'].count()

# Calculate the total number of accidents across all models
total_accidents = accidents_per_model.sum()

# Calculate the accident rate per aircraft model
accident_rate_per_model = accidents_per_model / total_accidents

# Display the accident rate per aircraft model
print(accident_rate_per_model.sort())
```
